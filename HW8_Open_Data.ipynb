```python
import requests as req
import pandas as pd
```

```python
remote_url = 'https://data.gov.ru/opendata/7303002440-karamzin2016/data-20160823T1237-structure-20160823T1237.csv?encoding=CP1251'
local_file = 'file.csv'
data = req.get(remote_url)
with open (local_file, 'wb') as file:
    file.write(data.content)
```

```python
df = pd.read_csv('file.csv', sep = ',', encoding = 'CP1251')
```
